# Import Things

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

# Read CSV

```python
df = pd.read_csv("C:\\Users\\acer\\WhiteIceAnalytics\\test.csv")
```

```python
df['True_Adj_Sv%'] = df['True_Adj_Sv%'].fillna(0)
df['GSAA'] = df['GSAA'].fillna(0)
```

```python
df['GP'] = df['GP'].astype(np.int64)
df['GS'] = df['GS'].astype(np.int64)
df['W'] = df['W'].astype(np.int64)
df['L'] = df['L'].astype(np.int64)
df['T'] = df['T'].astype(np.int64)
df['OT'] = df['OT'].astype(np.int64)
df['GA'] = df['GA'].astype(np.int64)
df['SO'] = df['SO'].astype(np.int64)
df['G'] = df['G'].astype(np.int64)
df['A'] = df['A'].astype(np.int64)
df['P'] = df['P'].astype(np.int64)
df['PIM'] = df['PIM'].astype(np.int64)
df['True_Adj_GAA'] = df['True_Adj_GAA'].round(2)
df['TOI'] = df['TOI'].round(2)
df['Actual GP'] = df['Actual GP'].round(2)
df['Prop GP'] = df['Prop GP'].round(2)
df['True_Adj_Sv%'] = df['True_Adj_Sv%'].round(3)
df['GSAA'] = df['GSAA'].round(2)
```

```python
df_25 = df.loc[df['25+'] == 'Y']
df_25 = df_25.drop(columns=['S/C', 'Actual GP', 'Prop GP', '25+'])
df_25 = df_25[['Player', 'Season', 'Team', 'GP', 'GS', 'True_Adj_GAA', 'True_Adj_Sv%', 'GSAA', 'W', 'L', 'T', 'OT', 'SO', 'GAA', 'Sv%', 'SA', 'Svs', 'GA', 'TOI', 'G', 'A', 'P', 'PIM']]
df_25 = df_25.rename(columns={'Player': 'Name', 'True_Adj_GAA': 'Adj. GAA', 'True_Adj_Sv%': 'Adj. Sv%'})
df_25 = df_25.reset_index(drop=True)
df_25['Season'] = df_25['Season'].astype(str)
df_25['Season1'] = df_25['Season'].map(lambda x: x[:4])
df_25['Season2'] = df_25['Season'].map(lambda x: x[4:])
for i in range(0, len(df_25)):
    df_25['Season'][i] = df_25['Season1'][i] + "-" + df_25['Season2'][i]
df_25 = df_25.drop(columns=['Season1', 'Season2'])
df_25
```

# Save As HTML

```python
df_25.to_html("C:\\Users\\acer\\WhiteIceAnalytics\\index.html", index = False)
```
